**LIST METHODS**

1.list.append(ele)

    adds an ele at the end of the list

2.list.pop(index)

    if index is given then removes the ele at that index
    index is optional
    if index is not mentioned,removes the last element
    returns popped ele when used with print func
3.list.remove(ele)

    removes the ele from the list
    ele is mandatory
    returns None when used with print func
4.list.insert(ind,ele)

    inserts the ele at the given ind
    returns None
5.list1.extend(list2)
 
    list2 elements will be added to list1
6.list.count(ele)

    returns the frequency of an ele
7.len(list)

    returns the length of the list
8.list2=list1.copy()

    copies list1 eles to list2
    changes made in list2 will not affect list1
9.list.reverse()

    reverse the eles of the list
10.list.sort(reverse=True)

    sorts the eles of the list in asc order by default
    reverse is optional
    if reverse=True,sorts in desc order
    performs operation on the same list
11.list2=sorted(list1,reverse=True)

    sorts list1 and stores in list2
12.list.clear()

    clears all the eles of a list



```python
l1=[1,3]
l1.append(4)
print("After appending:",l1)
l1.insert(1,2)
print("After inserting:",l1)
l1.pop()
print("After popping:",l1)
l1.remove(2)
print("After removing:",l1)
l1.sort()
print("After sorting:",l1)
l2=l1.copy()
print("After copying l1 to l2:",l2)
l2.clear()
print("After clearing l2:",l2)
print("l1:",l1)

```

**TYPE CASTING**

1.string to int

    int(string)
2.string to float
  
    float(string)
3.int to string

    str(integer)
4.string to list

    list(str)-returns a list of integer objects
5.list objects from str to other types

    list(map(type,list))
   
    

```python
#str to list
s='12345'
l1=list(s)
print(l1)
l2=list(map(int,l1))
print(l2)
```
