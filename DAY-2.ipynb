**TUPLE METHODS:**

1.tuple.index(element)

    returns the index of the first occurrence of the element
2.tuple.count(element)
 
    returns the frequency of an element

```python
t=(1,1,2,2,2)
print(t.index(2))
print(t.count(1))
```

**SET METHODS**

1.s1.union(s2)

    returns a set by combining both the sets
    original sets remains unchanged
2.s1.intersection(s2)
    
    returns the elements present in both the sets
    original sets remains unchanged
3.s1.difference(s2)
 
    removes the repeated elements and returns the remaining ele from first set
    original sets remains unchanged
4.s1.difference_update(s2)
 
    does not return anything just updates the first set with the result of difference function


```python
s1={1,2,3,4}
s2={3,4,5,6}
print("union=",s1.union(s2))
print("intersection=",s1.intersection(s2))
print("difference=",s1.difference(s2))
print("difference_update=",s1.difference_update(s2))
print("s1 after difference_update",s1)
  
```

**CONDITONAL STATEMENTS**

1.if

2.elif

3.else


```python
#if
n=2
if n%2==0:
  print("YES")
```

```python
#if else
num=int(input())
if num%2==0:
    print(num,"is even")
else:
    print(num,"is odd")
```

```python
#elif
n=int(input())
if n==1:
    print("MONDAY")
elif n==2:
    print("TUESDAY")
elif n==3:
    print("WEDNESDAY")
elif n==4:
    print("THURSDAY")
elif n==5:
    print("FRIDAY")
elif n==6:
    print("SATURDAY")
elif n==7:
    print("SUNDAY")
else:
    print("INVALID")
```

**DICTIONARY**

```python
#accessing values in a dict
d={1:'ANU',2:'BINDU',3:'PRASU',4:'TEJU',5:'VARSHI'}
print(d[3])
print(d[1])
```

**DICTIONARY METHODS**

1.d.get(key)
 
    for accessing or fetching the values
2.d.update({key-value})
    
    adds value to a dictionary
    takes a dictionary having a key-value pair to add
    can also update a value for a particular key
3.d.keys() 

    for accessing keys of a dictionary
4.d.values() 

    for accessing values of a dictionary

```python
d={}
d.update({'key1':1})
d.update({'key2':2,'key3':3})
print(d)
d.update({'key1':4})
print(d)
print(d.get('key1'))
print(d.keys())
print(d.values())
```
