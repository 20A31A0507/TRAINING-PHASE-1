```python
#2D matrix addition
r=3
c=3
arr1=[]
for i in range(r):
    l=list(map(int,input().split()))[:c]
    arr1.append(l)
print(arr1)

arr2=[]
for j in range(r):
    l=list(map(int,input().split()))[:c]
    arr2.append(l)
print(arr2)

res=[[0 for j in range(c)] for i in range(r)]
for i in range(r):
    for j in range(c):
        res[i][j]=arr1[i][j]+arr2[i][j]
        print(res[i][j],end=' ')
    print()
```

```python
#method-1:
def prime(num):
    count=0
    for i in range(1,num+1): 
        if num%i==0:
            count+=1
    if count==2:
        print(num,"is a prime number")
    else:
        print(num,"is not a prime number") 
x=int(input())
prime(x)
```

```python
#method-2:
def prime(num):
    count=0
    for i in range(2,num): 
        if num%i==0:
            count+=1
    if count==0:
        print(num,"is a prime number")
    else:
        print(num,"is not a prime number") 
x=int(input())
prime(x)
```

```python
#method-3:
def prime(num):
    count=0
    for i in range(2,num//2): 
        if num%i==0:
            count+=1
    if count==0:
        print(num,"is a prime number")
    else:
        print(num,"is not a prime number") 
x=int(input())
prime(x)
```

```python
#method-4:
def prime(num):
    count=0
    for i in range(2,int(num**0.5)+1): 
        if num%i==0:
            count+=1
    if count==0:
        print(num,"is a prime number")
    else:
        print(num,"is not a prime number") 
x=int(input())
prime(x)
```

4.VARIABLE LENGTH ARGUMENT FUNC
    
    we can pass any no. of argu
    always put variable length argument at the end of the arg list

```python
#REGULAR FUNC
def func(num1,num2):
  print(num1,num2)
a,b=10,20
func(a,b)
```

```python
#DEFAULT ARGUMENT FUNC
def func(num1=0,num2=0):
    print(num1,num2)
a,b=10,20
func(a)
  
```

```python
#KEYWORD ARGUMENT FUNCTIONS
def func(num1=0,num2=0):
    print(num1,num2)
func(num2=40,num1=20)
```

```python
#VARIABLE LENGTH ARGUMENT FUNC
def func(*abc):
  print(sum(abc))
func(1,2,3)
   
```

```python
#RECURSION
def check(n):
  print(n,end=" ")
  if n>0:
    check(n-1)
check(5)
```
