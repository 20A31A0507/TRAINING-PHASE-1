**INHERITANCE**

Types:

1.Single

    one parent-one child class 

2.Multilevel

    one parent class-one child class-subchild class

3.Hierarchical

    one parent class-multiple child classes

4.Multiple

    multiple parent classes-one child class

5.Hybrid

    combination of more than one type

```python
#SINGLE
class A:
    name="abc"
class B(A):
    pass
obj=B()
print(obj.name)
```

```python
#MULTILEVEL
class Birds:
    prop1="can fly"
    prop2="cannot fly"
class bird1(Birds):
    name1="parrot"
    prop3="has two wings"
class bird2(bird1):
    name2="ostrich"
obj=bird2()
print(obj.name2,obj.prop3,"but",obj.prop2)
print(obj.name1,obj.prop3,"and",obj.prop1)
```

```python
#HIERARCHICAL
class vehicles:
    color1="yellow"
    color2="white"
    color3="black"
    wheel1=4
    wheel2=3
    wheel3=2
class v1(vehicles):
    v_name="Bus"
class v2(vehicles):
    v_name="Auto"
class v3(vehicles):
    v_name="Bike"

obj1=v1()
print("name:",obj1.v_name)
print("color:",obj1.color1)
print("no. of wheels:",obj1.wheel1)
print("\n")
obj2=v2()
print("name:",obj2.v_name)
print("color:",obj2.color1)
print("no. of wheels:",obj2.wheel2)
print("\n")
obj3=v3()
print("name:",obj3.v_name)
print("color:",obj3.color3)
print("no. of wheels:",obj3.wheel3)
```

```python
#MULTIPLE
class father:
    height=5.7
class mother:
    color="white"
    height=5.3
class child(father,mother): #BOTH MOTHER ND FATHER HAD HEIGHT VARIABLE, BUT VARIABLE OF THE FIRST PASSED CLASS IS TAKEN
    pass
obj=child()
print("child height=",obj.height)
print("child color=",obj.color)
```

```python
#HYBRID
class movies:
    type1="Biography"
    type2="Comedy"
    type3="Thriller"
class movie1(movies):
    movie="Mahanati"
class movie2(movies):
    movie="Hit-2"
class movie_type(movie1,movie2):
    rating=5 
obj=movie_type()
print("Movie:",obj.movie,"| Type:",obj.type1,"| Rating=",obj.rating)
```

```python
#ROCK-PAPER-SCISSORS WITH SYSTEM
import random
l=['rock','paper','scissors']
user_score=0
sys_score=0
while user_score!=5 and sys_score!=5:
    user_ch=input()
    if user_ch not in l:
        print("Invalid Input")
        print("\n")
    else:
        x=random.randint(0,2)
        sys_ch=l[x]
        print(sys_ch)
        if sys_ch==user_ch:
            pass
        elif (sys_ch=='rock' and user_ch=='paper') or (sys_ch=='paper' and user_ch=='scissors') or (sys_ch=='scissors' and user_ch=='rock'):
            user_score+=1
        else:
            sys_score+=1
        print("your score=",user_score,"system's_score=",sys_score)
        if user_score>sys_score:
            print("YOU ARE LEADING BY",(user_score-sys_score))
        elif sys_score>user_score:
            print("SYSTEM IS LEADING BY",(sys_score-user_score))
        else:
            print("BOTH ARE SAME")
        print("\n")
print("==================================")   
print("YOUR FINAL SCORE:",user_score)
print("SYSTEM'S FINAL SCORE:",sys_score)
print("==================================")
if user_score>sys_score:
    print("CONGRATULATIONS!YOU WON THE GAME")
elif sys_score>user_score:
    print("SYSTEM WON THE GAME")
else:
    pass
```

**POLYMORPHISM**

same method behaving differently in different situations

1.runtime polymorphism or method overriding

2.compiletime polymorphism or method overloading


```python
#METHOD OVERRIDING OR RUNTIME POLYMORPHISM
class A:
    def method_1(self,a,b):   
        print("sum=",a+b)
class B(A):
    def method_1(self,a,b):   
        print("product=",a*b)
obj=B()
obj.method_1(10,20) #SINCE OBJ IS CREATED FOR CLASS B, THE METHOD IN CLASS B WILL BE CHECKED FIRST
```

```python
#METHOD OVERLOADING OR COMPILETIME POLYMORPHISM
class A:
    def method_1(self,a,b):   
        print("sum=",a+b)
class B(A):
    def method_1(self,a,b):   #m1
        print("product=",a*b) 
    def method_1(self,abc):   #m2
        print("value=",abc)   
obj=B()
obj.method_1(100)      #SINCE m2 OVERLOADS m1, WE CAN GIVE ONLY ONE ARGUMENT AS THE METHOD WILL BE REPLACED
```
