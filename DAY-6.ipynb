```python
class Animals:
    def speak(self,animal):
        print(animal,"can speak")
class anim1(Animals):
    name='Dog'
class anim2(Animals):
    name='Cat'
class anim3(Animals):
    name='Cow'
obj1=anim1()
obj1.speak(obj1.name)
obj2=anim2()
obj2.speak(obj2.name)
obj3=anim3()
obj3.speak(obj3.name)
```

from abc import ABC,abstractmethod------abc-abstract class nd ABC-abstract base class

```python
from abc import ABC,abstractmethod    
class Area(ABC):
    @abstractmethod
    def square_area(self):
        pass
    @abstractmethod
    def rec_area(self):
        pass
class square(Area):
    def square_area(self):
        print("Square area=side*side")
    def rec_area(self):
        pass
class rectangle(Area):
    def square_area(self):
        pass
    def rec_area(self):
        print("Rectangle area=2(lengtgh+breadth)")
obj1=square()
obj1.square_area()
obj2=rectangle()
obj2.rec_area()
```

```python
#abstraction-example-2
from abc import ABC,abstractmethod    
class features(ABC):
    @abstractmethod
    def ostrich_features(self):
        pass
    @abstractmethod
    def parrot_features(self):
        pass
class ostrich(features):
    def ostrich_features(self):
        print("Ostrich is a bird but cannot fly")
    def parrot_features(self):
        pass
class parrot(features):
    def ostrich_features(self):
        pass
    def parrot_features(self):
        print("Parrot is a bird but can fly")
obj1=ostrich()
obj1.ostrich_features()
obj2=parrot()
obj2.parrot_features()
```
